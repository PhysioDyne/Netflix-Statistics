Finding the most frequent players in each cast's strings

```python

def func():
    memoizate = {}
    df["cast"].fillna(" ",inplace=True)
    for i in range(1,df["cast"].count()):
        castlist = df["cast"][i].split(", ")
        for j in castlist:
            if j == " ":
                continue
            if j not in memoizate:
                memoizate[j] = 1
            else:
                memoizate[j]+= 1
    for i in memoizate:
        if memoizate[i] == max(memoizate.values()):
            return i
```

```python
func()
```

Average Duration of All Series and Films

```python

df["duration"].fillna("0",inplace=True)
durations = (df[df["duration"].apply(lambda x: "Season" not in x)]["duration"].apply(lambda x: x.split(" ")[0]))
```
